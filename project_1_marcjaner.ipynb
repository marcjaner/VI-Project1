## Data cleaning

In this section we will be working with the raw data obtained from the three dataframes in order to clean it and come up with the final datasets that will be later used to create the visualizations

### Motor vehicle collisions

#### Auxiliary functions

This section contains a set of auxiliary functions that will be used througout the data cleaning process.

```python
def accidents_remove_unwanted_columns(df, WANTED_COLUMNS):
    df = df[WANTED_COLUMNS]
    df.head()

    return(df)

    
```

```python
# Load the df
vc_raw = pd.read_csv('./Motor_Vehicle_Collisions-RAW.csv')

print(vc_raw.head())
```

#### Extract 2018 summer data

```python
vc_raw['CRASH DATE'] = pd.to_datetime(vc_raw['CRASH DATE'])

vc_2018_raw = vc_raw[(vc_raw['CRASH DATE'] >= '2018-06-01') & (vc_raw['CRASH DATE'] <= '2018-09-30')]

vc_2018_raw.head()
```

#### Extract 2020 summer data

```python
vc_raw['CRASH DATE'] = pd.to_datetime(vc_raw['CRASH DATE'])

vc_2020_raw = vc_raw[(vc_raw['CRASH DATE'] >= '2020-06-01') & (vc_raw['CRASH DATE'] <= '2020-09-30')]

print(vc_2020_raw.columns)
```

#### Keep only wanted columns

In order to reduce the number of dimension we will be working with, we will eliminate those columns that will not be needed at the early stages of the data cleaning process.

```python
# Define wanted columns
WANTED_COLUMNS = [
    'CRASH DATE',
    'CRASH TIME',
    'BOROUGH',
    'ZIP CODE',
    'LATITUDE',
    'LONGITUDE',
    'NUMBER OF PERSONS INJURED',
    'NUMBER OF PERSONS KILLED', 
    'NUMBER OF PEDESTRIANS INJURED',
    'NUMBER OF PEDESTRIANS KILLED', 
    'NUMBER OF CYCLIST INJURED',
    'NUMBER OF CYCLIST KILLED', 
    'NUMBER OF MOTORIST INJURED',
    'NUMBER OF MOTORIST KILLED',
    'CONTRIBUTING FACTOR VEHICLE 1',
    'VEHICLE TYPE CODE 1'
]


vc_2018_raw = accidents_remove_unwanted_columns(vc_2018_raw, WANTED_COLUMNS)
vc_2020_raw = accidents_remove_unwanted_columns(vc_2020_raw, WANTED_COLUMNS)
```

#### Modify column names
This allows us to erase spaces from the columns names in order to avoid having problems in the future

```python
# Remove ' ' from column names. Replace them by '_'
vc_2018_raw.columns = [col.replace(' ', '_') for col in vc_2018_raw.columns]
vc_2020_raw.columns = [col.replace(' ', '_') for col in vc_2020_raw.columns]

```

#### Dealing with missing values

```python
#Check missing values

vc_2018_nan = vc_2018_raw.isna().sum()
print(vc_2018_nan)

vc_2020_nan = vc_2020_raw.isna().sum()
print(vc_2020_nan)

```

```python
#Remove rows with nan in columns LATITUDE, LONGITUDE, CONTRIBUTING_FACTOR_VEHICLE_1 and VEHICLE_TYPE_CODE_1

vc_2018_nan = vc_2018_raw.dropna(subset=['LATITUDE', 'LONGITUDE', 'CONTRIBUTING_FACTOR_VEHICLE_1', 'VEHICLE_TYPE_CODE_1'])
print(vc_2018_nan.isna().sum())

vc_2020_nan = vc_2020_raw.dropna(subset=['LATITUDE', 'LONGITUDE', 'CONTRIBUTING_FACTOR_VEHICLE_1', 'VEHICLE_TYPE_CODE_1'])
```

### Weather data

#### Auxiliary functions

Remove attribute columns

```python
def remove_attribute_columns(df):
    columns_to_drop = [col for col in df.columns if 'ATTRIBUTES' in col]
    
    return (df.drop(columns=columns_to_drop))
```
